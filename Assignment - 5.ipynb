```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
```

```python
# data from the downloadable file
y = [2.12, 2.71, 3.44 ,2.76, 2.72, 0.96 ,2.00 ,3.26 ,2.50,
1.20 ,1.62 ,1.96 ,2.60 , 1.30 ,2.67 ,4.40 ,1.80 ,4.90 ,2.39,
1.62 ,1.47 ,0.89, 2.52, 1.21, 0.90, 0.80]

log_y = np.log(y)

x = np.linspace(-2, 10, 200)
```

```python
shapiro_density = sp.shapiro(y)
shapiro_log_density = sp.shapiro(log_y)
```

```python
mean,sigma = sp.norm.fit(y)
dist = sp.norm(mean,sigma)
```

```python
mean_,sigma_ = sp.norm.fit(log_y)
dist_ = sp.norm(mean_,sigma_)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].hist(y,bins=9,density=True)
ax[0].plot(x,dist.pdf(x))
ax[0].set_xlabel('Density')
ax[0].set_ylabel('Number of asteroids')

ax[1].hist(log_y,bins=9,density=True)
ax[1].plot(x,dist_.pdf(x))
ax[1].set_xlabel('log(density)')
ax[1].set_ylabel('number of astroids')

plt.show()
```
